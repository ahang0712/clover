```python
import os
import json

# 指定目录路径
directory = '/home/vip/hanghe/race-master/llvm-api-demo/c-src/SV-COMP-2022-Benchmark'

def count_global_var_in_json_files(directory):
    total_count = 0
    file_counts = {}  # 用于记录每个文件的 GLOBAL_VAR 计数

    # 遍历指定目录及其子目录下所有文件
    for root, dirs, files in os.walk(directory):
        for filename in files:
            # 如果是 .json 文件
            if filename.endswith('.json'):
                file_path = os.path.join(root, filename)
                
                try:
                    # 打开并读取 JSON 文件
                    with open(file_path, 'r') as f:
                        data = json.load(f)
                        
                        # 统计每个文件中的 GLOBAL_VAR
                        file_count = 0
                        if 'GLOBAL_VAR' in data:
                            global_var = data['GLOBAL_VAR']
                            if isinstance(global_var, int):
                                file_count = global_var
                            elif isinstance(global_var, list):
                                file_count = len(global_var)

                        # 将每个文件的计数记录下来
                        file_counts[file_path] = file_count
                        total_count += file_count
                except (json.JSONDecodeError, FileNotFoundError) as e:
                    print(f"Error reading {filename}: {e}")

    return total_count, file_counts

# 调用函数并打印总和以及每个文件的计数
total_global_var_count, file_counts = count_global_var_in_json_files(directory)

print(f"Total 'GLOBAL_VAR' count across all .json files: {total_global_var_count}")
print("Individual file counts:")
for file_path, count in file_counts.items():
    print(f"{file_path}: {count}")
```
